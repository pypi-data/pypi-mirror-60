```python
import numpy as np
import matplotlib.pyplot as P
%matplotlib inline

import gwsurrogate
```

```python
sur = gwsurrogate.LoadSurrogate('PEOB')
```

```python
q = 1.5
chiA = [-0.2, 0.4, 0.1]
chiB = [-0.5, 0.2, -0.4]
dt = 0.1        # timestep size, Units of M
f_low = 0       # initial frequency, f_low=0 returns the full surrogate
t, h, dyn = sur(q, chiA, chiB, dt=dt, f_low=f_low, ellMax=2)   # dyn stands for dynamics, do dyn.keys() to see contents
```

```python
P.plot(t, h[(2,2)].real, label='l2m2 real')
P.plot(t, h[(2,1)].real, label='l2m1 real')
#P.plot(t, h[(3,3)].real, label='l3m3 real')
P.ylabel('Re[$h_{lm}$]', fontsize=18)
P.xlabel('t [M]', fontsize=18)
P.legend()
```
